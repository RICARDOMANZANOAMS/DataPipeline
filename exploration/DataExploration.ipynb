```python
import pandas as pd
import glob as glob
pathDataFolder="../data/*.csv"  #Define path to the directory with data 
pathFiles=glob.glob(pathDataFolder) #Get file names from directory
print(pathFiles)  #print path files
dfs=[]  #Array to concatenate pandas df one by one
for pathFile in pathFiles:  #Iterate through each path
    dfs.append(pd.read_csv(pathFile)) #Append each file into one array
dfAll=pd.concat(dfs) #Concatenate all pandas dataframe in the array into a single 
print(dfAll) 
```

```python
#Statistics
#Get name of the columns
print(f'Data Columns: {dfAll.columns}') # List the column names
```

```python
#Get frequency of classes
classesFreq=dfAll['Label_34_class'].value_counts()
print(f'Classes frequency: {classesFreq}')
```

```python
#Get null rows
nullRows=dfAll[dfAll.isnull().any(axis=1)]
print(f'Rows with nulls {nullRows}')
```

```python
#Get rows with Inf
import numpy as np

numericColumns= dfAll.select_dtypes(include=[np.number])
infMask = np.isinf(numericColumns).any(axis=1)
rowsInf=dfAll[infMask]
print(rowsInf)
```
